```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("diamonds.csv")

df
```

```python
df.columns
```

```python
#cleaning

df.isna().sum()
```

```python
df['Cut'].unique()
```

```python
df['Cut'].fillna('NA',inplace=True)
df['Fluorescence'].fillna('NA',inplace=True)
df['Culet'].fillna('NA',inplace=True)
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df
```

```python
df.rename(columns={'Carat Weight':'Carat_Weight', 'Length/Width Ratio': 'LWRatio','Depth %':'Depth','Table %':'Table'},inplace=True)

df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#encode all non numeric columns to numeric in df2
df2=df.copy()
from sklearn.preprocessing import LabelEncoder

shape_encoder=LabelEncoder()
shape_encoder.fit(df2['Shape'])

cut_encoder=LabelEncoder()
cut_encoder.fit(df2['Cut'])

color_encoder=LabelEncoder()
color_encoder.fit(df2['Color'])

clarity_encoder=LabelEncoder()
clarity_encoder.fit(df2['Clarity'])

polish_encoder=LabelEncoder()
polish_encoder.fit(df2['Polish'])

symmetry_encoder=LabelEncoder()
symmetry_encoder.fit(df2['Symmetry'])

girdle_encoder=LabelEncoder()
girdle_encoder.fit(df2['Girdle'])

culet_encoder=LabelEncoder()
culet_encoder.fit(df2['Culet'])

type_encoder=LabelEncoder()
type_encoder.fit(df2['Type'])

fluorescence_encoder=LabelEncoder()
fluorescence_encoder.fit(df2['Fluorescence'])

df2['Shape']=shape_encoder.transform(df2['Shape'])
df2['Cut']=cut_encoder.transform(df2['Cut'])
df2['Color']=color_encoder.transform(df2['Color'])
df2['Clarity']=clarity_encoder.transform(df2['Clarity'])
df2['Polish']=polish_encoder.transform(df2['Polish'])
df2['Symmetry']=symmetry_encoder.transform(df2['Symmetry'])
df2['Girdle']=girdle_encoder.transform(df2['Girdle'])
df2['Culet']=culet_encoder.transform(df2['Culet'])
df2['Type']=type_encoder.transform(df2['Type'])
df2['Fluorescence']=fluorescence_encoder.transform(df2['Fluorescence'])
```

```python
df2.head()
```

```python
#correlation matrix
corr = df2.corr()
plt.figure(figsize=(15,10))
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
x=df2.drop(['Price'],axis=1)
y=df2['Price']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)

x_train_original=x_train.copy()
x_test_original=x_test.copy()
y_train_original=y_train.copy()
y_test_original=y_test.copy()
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lr_model=LinearRegression()
lr_model.fit(x_train,y_train)

y_pred=lr_model.predict(x_test)
accuracy=lr_model.score(x_test,y_test)
print("Accuracy of Linear Regression model: ",accuracy)

rmse=np.sqrt(mean_squared_error(y_test,y_pred))
print("Root Mean Squared Error of Linear Regression model: ",rmse)
```

```python
#box plot for price
plt.figure(figsize=(10,10))
sns.boxplot(y='Price',data=df2)
```

```python
df3=df2.query('Price<=8500')
df3.shape
```

```python
#box plot for price
plt.figure(figsize=(10,10))
sns.boxplot(y='Price',data=df3)
```

```python
x=df3.drop(['Price'],axis=1)
y=df3['Price']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)

lr_model2=LinearRegression()
lr_model2.fit(x_train,y_train)

y_pred=lr_model2.predict(x_test)
accuracy=lr_model2.score(x_test,y_test)
print("Accuracy of Linear Regression model for price<=10000: ",accuracy)

rmse=np.sqrt(mean_squared_error(y_test,y_pred))
print("Root Mean Squared Error of Linear Regression model for price<=10000: ",rmse)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression

lr_model3=LogisticRegression()
lr_model3.fit(x_train,y_train)
y_pred=lr_model3.predict(x_test)
accuracy=lr_model3.score(x_test,y_test)
print("Accuracy of Logistic Regression model: ",accuracy)

rmse=np.sqrt(mean_squared_error(y_test,y_pred))
print("Root Mean Squared Error of Logistic Regression model: ",rmse)
```

```python
#bayesian ridge regression
from sklearn.linear_model import BayesianRidge

br_model=BayesianRidge()
br_model.fit(x_train,y_train)
y_pred=br_model.predict(x_test)
accuracy=br_model.score(x_test,y_test)

print("Accuracy of Bayesian Ridge Regression model: ",accuracy)

rmse=np.sqrt(mean_squared_error(y_test,y_pred))
print("Root Mean Squared Error of Bayesian Ridge Regression model: ",rmse)
```

```python
#bayesian ridge regression
from sklearn.linear_model import BayesianRidge

br_model=BayesianRidge()
br_model.fit(x_train_original,y_train_original)
y_pred=br_model.predict(x_test_original)
accuracy=br_model.score(x_test_original,y_test_original)

print("Accuracy of Bayesian Ridge Regression model: ",accuracy)

rmse=np.sqrt(mean_squared_error(y_test_original,y_pred))
print("Root Mean Squared Error of Bayesian Ridge Regression model: ",rmse)
```

```python
#tune model linear regression

from sklearn.model_selection import GridSearchCV

param_grid={
    'fit_intercept':[True,False],
    'copy_X':[True,False]
}

grid=GridSearchCV(lr_model2,param_grid,verbose=3)
grid.fit(x_train,y_train)

grid.best_params_
```

```python
print("Best parameters for Linear Regression model: ",grid.best_params_)
print("Best score for Linear Regression model: ",grid.best_score_)
print("Best estimator for Linear Regression model: ",grid.best_estimator_)
```

```python
#get best model mse
y_pred=grid.best_estimator_.predict(x_test)
rmse=np.sqrt(mean_squared_error(y_test,y_pred))
print("Root Mean Squared Error of Linear Regression model: ",rmse)
```
