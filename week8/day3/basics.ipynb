```python
import matplotlib.pyplot as plt


x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y)
```

```python
# proper way to show the plot
ax,fig = plt.subplots()
fig.plot(x,y)

fig.set_title("My first plot")

fig.set_xlabel("X-axis")
fig.set_ylabel("Y-axis")
plt.show()

```

```python
# proper way to show the plot
ax,fig = plt.subplots()
fig.scatter(x,y)

fig.set_title("My first plot")

fig.set_xlabel("X-axis")
fig.set_ylabel("Y-axis")
plt.show()
```

```python
ax.savefig("my_first_plot.png")
```

```python
# proper way to show the plot
ax,fig = plt.subplots()
fig.plot(x,y,marker='o',color='red',linestyle='dotted',label="Profits")
fig.legend() #display the label
fig.set_title("My first plot")

fig.set_xlabel("X-axis")
fig.set_ylabel("Y-axis")
plt.show()
```

```python
# proper way to show the plot
ax,fig = plt.subplots()
fig.plot(x,y,marker='o',color='red',linestyle='dotted',label="Profits")
fig.legend() #display the label
fig.set_title("My first plot")

#annotation 
fig.annotate("Good profit",xy=(3,9),xytext=(2.0,13),arrowprops=dict(facecolor='black',shrink=0.05))

fig.set_xlabel("X-axis")
fig.set_ylabel("Y-axis")
plt.show()
```

```python
a=[1,2,3,4,5]
b=[1,4,9,16,25]

x=[1,2,3,4,5]
y=[25,16,9,4,1]

ax, fig=plt.subplots()

fig.plot(a,b,marker='o',color='red',linestyle='dotted',label="Profits")
fig.plot(x,y,marker='x',color='blue',linestyle='dotted',label="Loss")
fig.legend() #display the label

fig.set_title("Profit and loss")
fig.set_xlabel("Day")
fig.set_ylabel("Total")

plt.show()
```
